# Lecture Demos - Fourier Series

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def fs_ex2a():
    eps = 1.e-4
    n = 1001
    T = 4.*np.pi
    x = np.linspace(-6*np.pi+eps,6*np.pi-eps,n)   
    f = np.zeros(n)
    for i in range(n):
        xc = x[i]-np.rint(x[i]/T)*T
        if -2.*np.pi <= xc <= 0.:
            f[i] = -1.*xc
        else:
            f[i] = 1.*xc
    plotfs(x,f)

def fs_ex2b():
    eps = 1.e-4
    n = 1001
    T = 2.*np.pi
    x = np.linspace(-3.*np.pi+eps,3.*np.pi-eps,n)   
    f = np.zeros(n)
    for i in range(n):
        xc = x[i]-np.rint(x[i]/T)*T
        f[i] = np.sin(x[i])
    plotfs(x,f)

def fs_cos():
    eps = 1.e-4
    n = 1001
    T = 2.*np.pi
    x = np.linspace(-3.*np.pi+eps,3.*np.pi-eps,n)   
    f = np.zeros(n)
    for i in range(n):
        xc = x[i]-np.rint(x[i]/T)*T
        f[i] = np.cos(x[i])
    plotfs(x,f)
    
def fs_ex2c():
    eps = 1.e-4
    n = 1001
    T = 2
    x = np.linspace(-3.+eps,3.-eps,n)   
    f = np.zeros(n)
    for i in range(n):
        xc = x[i]-np.rint(x[i]/T)*T
        if 0. <= xc <= 1.:
            f[i] = 2.
        else:
            f[i] = 1.
    plotfs(x,f)   

def fs_ex5a_sine(disp_psum, sn):
    eps = 1.e-4
    n = 1001
    T = 2.
    x = np.linspace(-3+eps,3-eps,n)   
    f = np.zeros(n)
    for i in range(n):
        xc = x[i]-np.rint(x[i]/T)*T
        f[i] = xc
    psum = np.zeros(n)
    if disp_psum:
        for i in range(sn):
            psum = psum-(2.*np.cos((i+1)*np.pi)*np.sin((i+1)*x*np.pi))/((i+1)*np.pi)
    plotfs(x,f, psum, disp_psum)

def fs_ex5a_cosine(disp_psum, sn):
    eps = 1.e-4
    n = 1001
    T = 2.
    x = np.linspace(-3+eps,3-eps,n)   
    f = np.zeros(n)
    for i in range(n):
        xc = x[i]-np.rint(x[i]/T)*T
        if -1 <= xc < 0.:
            f[i] = -1.*xc
        else:
            f[i] = xc
    psum = np.zeros(n)
    psum += 0.5 # add a0 coefficient
    if disp_psum:
        for i in range(sn):
            psum = psum+(2.*(np.cos((i+1)*np.pi)-1.)*np.cos((i+1)*x*np.pi))/((i+1)*(i+1)*np.pi*np.pi)
    plotfs(x,f, psum, disp_psum)

def fs_ex5b_sine(disp_psum, sn):
    eps = 1.e-4
    n = 1001
    T = 2.
    x = np.linspace(-3+eps,3-eps,n)   
    f = np.zeros(n)
    for i in range(n):
        xc = x[i]-np.rint(x[i]/T)*T
        if -1 <= xc < 0.:
            f[i] = -1.
        else:
            f[i] = 1.
    psum = np.zeros(n)
    if disp_psum:
        for i in range(sn):
            psum = psum-(2.*(np.cos((i+1)*np.pi)-1.)*np.sin((i+1)*x*np.pi))/((i+1)*np.pi)
    plotfs(x,f, psum, disp_psum)
    
def fs_ex5b_cosine(disp_psum, sn):
    eps = 1.e-4
    n = 1001
    T = 4.
    x = np.linspace(-6+eps,6-eps,n)   
    f = np.zeros(n)
    for i in range(n):
        xc = x[i]-np.rint(x[i]/T)*T
        if -1. <= xc < 1.:
            f[i] = 1.
        else:
            f[i] = 0.
    psum = np.zeros(n)
    psum += 0.5
    if disp_psum:
        for i in range(sn):
            psum = psum+(2.*np.sin((i+1)*np.pi/2.)*np.cos((i+1)*x*np.pi/2.))/((i+1)*np.pi)
    plotfs(x,f, psum, disp_psum)
    
def fs_ex7a(disp_psum, sn):
    eps = 1.e-4
    n = 1001
    T = 2.*np.pi
    x = np.linspace(-3*np.pi+eps,3*np.pi-eps,n)   
    f = np.zeros(n)
    for i in range(n):
        xc = x[i]-np.rint(x[i]/T)*T
        if -1*np.pi <= xc < -0.5*np.pi:
            f[i] = -1.
        elif -0.5*np.pi <= xc < 0.5*np.pi:
            f[i] = 1.
        elif 0.5*np.pi <= xc <= 1*np.pi:
            f[i]=-1.
    psum = np.zeros(n)
    if disp_psum:
        for i in range(sn):
            psum = psum+4.*np.sin((i+1)*np.pi/2.)*np.cos((i+1)*x)/((i+1)*np.pi)
    plotfs(x,f, psum, disp_psum)

def fs_ex7b(disp_psum, sn):
    eps = 1.e-4
    n = 1001
    T = 2.*np.pi
    x = np.linspace(-3*np.pi+eps,3*np.pi-eps,n)   
    f = np.zeros(n)
    for i in range(n):
        xc = x[i]-np.rint(x[i]/T)*T
        if -1*np.pi <= xc < -0.5*np.pi:
            f[i] = -1.*xc-np.pi
        elif -0.5*np.pi <= xc < 0.5*np.pi:
            f[i] = xc
        elif 0.5*np.pi <= xc <= 1*np.pi:
            f[i]= -1.*xc+np.pi
    psum = np.zeros(n)
    if disp_psum:
        for i in range(sn):
            psum = psum+4.*np.sin((i+1)*np.pi/2.)*np.sin((i+1)*x)/((i+1)*(i+1)*np.pi)
    plotfs(x,f, psum, disp_psum)


def plotfs(x,f, psum = 0, disp_psum = False):
    fig = plt.figure(figsize=(16, 8))
    ax1 = fig.add_subplot(111)
    ax1.set_xlabel('x')
    ax1.set_ylabel('f(x)')
    ax1.plot(x,f, 'k-')
    ax1.axhline(linewidth=0.5, color='k')
    ax1.axvline(linewidth=0.5, color='k')
    if disp_psum:
        ax1.plot(x,psum, 'r--')
    plt.show()
```
