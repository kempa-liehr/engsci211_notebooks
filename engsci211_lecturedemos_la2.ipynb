# Lecture Demos - Linear Algebra 2

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import linalg
%matplotlib inline

def showeigenstuff(A, vec, showevectors = True):
    w, v = linalg.eig(A)
    a = 1./np.sqrt(2) # basic length of box
    orig_x = np.array([-1.*a,1.*a,1.*a,-1.*a,-1.*a])
    orig_y = np.array([-1.*a,-1.*a,1.*a,1.*a,-1.*a])
    deform_x = np.copy(orig_x)
    deform_y= np.copy(orig_y)
    for i in range(len(orig_x)):
        orig = np.array([orig_x[i],orig_y[i]])
        deform = np.matmul(A,orig)
        deform_x[i] = deform[0]
        deform_y[i] = deform[1]
    eline1 = np.matmul(A,v[:,0])
    eline2 = np.matmul(A,v[:,1])
    vline = np.matmul(A,vec)

    fig = plt.figure(figsize=(16, 8))
    ax1 = fig.add_subplot(111)
    ax1.set_xlabel('x')
    ax1.set_ylabel('y')
    ax1.axhline(linewidth=1, color='k')
    ax1.axvline(linewidth=1, color='k')
    ax1.plot(orig_x, orig_y, 'k--o', label='Original Shape')
    ax1.plot(deform_x, deform_y, 'g--o', label='Deformed Shape')
    ax1.plot([0.,vec[0]],[0.,vec[1]] , 'm-', label='Original vector')
    ax1.plot([0.,vline[0]],[0.,vline[1]] , 'm--', label='Deformed vector')
    if showevectors:
        ax1.plot([-1.*eline1[0],eline1[0]], [-1.*eline1[1],eline1[1]], 'r-', label=r'$\lambda_1 \vec{s}_1$')
        ax1.plot([-1.*eline2[0],eline2[0]], [-1.*eline2[1],eline2[1]], 'b-', label=r'$\lambda_2 \vec{s}_2$')
    ax1.legend(fontsize=12)
```

```python
# Example 5.1
A = np.array([[5, -2], [1, 2]])

# Example 5.2
#A = np.array([[5, -2], [-2, 2]])

# Additional example
#A = np.array([[1.5, 0.25], [0.25, 1.5]])

showeigenstuff(A, vec = [0.3,0.6], showevectors = False)
```
