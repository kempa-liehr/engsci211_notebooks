# Linear Algebra - Iterative Methods

```python
# Initialisation of functions used in lectures
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def iterative_1(x0=0, y0=0, method_j=True, method_gs=False, disp=True, disp_j=False, disp_gs=False, convergent=True):
    # define system of equations
    x1 = np.linspace(0.,1.5,1001)
    y1 = 2.*x1-1.
    x2 = np.linspace(0.,2.,1001)
    y2 = (3.-x2)/2.
    tol = 5.e-5

    if disp:
        fig = plt.figure(figsize=(16, 8))
        ax1 = fig.add_subplot(111)
        #ax1.set_xlim([-0.5, 2.5])
        #ax1.set_ylim([-1.5, 2.5])
        ax1.set_xlabel('x')
        ax1.set_ylabel('y')
        ax1.axhline(linewidth=2, color='k')
        ax1.axvline(linewidth=2, color='k')
        ax1.plot(1,1, 'ko')
        line1 = ax1.plot(x1,y1, 'r-', label='2x-y=1')
        line2 = ax1.plot(x2,y2, 'b-', label='x+2y=3')
        ax1.legend(fontsize=20)    
    if method_j:
        xj = [x0]
        yj = [y0]
        print("Jacobi Method\nIteration xold    yold    xnew    ynew")
        for i in range(20):
            if convergent:
                xj.append(0.5*yj[i]+0.5)
                yj.append(-0.5*xj[i]+1.5)
            else:
                xj.append(3.-2*yj[i])
                yj.append(2.*xj[i]-1.)
            print("%02d        %6.4f  %6.4f  %6.4f  %6.4f" % (i+1,xj[i],yj[i],xj[i+1],yj[i+1]))
            if np.abs(xj[i+1]-1.) <= tol and np.abs(yj[i+1]-1.) <= tol:
                print('\n')
                break                
    if method_gs:
        xgs = [x0]
        ygs = [y0]
        print("Gauss-Seidel Method\nIteration xold    yold    xnew    ynew")
        for i in range(20):
            if convergent:
                xgs.append(0.5*ygs[i]+0.5)
                ygs.append(-0.5*xgs[i+1]+1.5)
            else:
                xgs.append(3.-2*ygs[i])
                ygs.append(2.*xgs[i+1]-1.)            
            print("%02d        %6.4f  %6.4f  %6.4f  %6.4f" % (i+1,xgs[i],ygs[i],xgs[i+1],ygs[i+1]))
            if np.abs(xgs[i+1]-1.) <= tol and np.abs(ygs[i+1]-1.) <= tol:
                print('\n')
                break
    if disp_j:
        line_j = ax1.plot(xj,yj, 'g--o', label='Jacobi Method')
        ax1.legend(fontsize=20)
    if disp_gs:
        line_gs = ax1.plot(xgs,ygs, 'm--o', label='Gauss-Seidel Method')
        ax1.legend(fontsize=20)
        
def iterative_2(x1_i=0, x2_i=0, x3_i=0, method_j=True, method_gs=False):
    tol = 5.e-5
    if method_j:
        x1_j = [x1_i]
        x2_j = [x2_i]
        x3_j = [x3_i]
        print("Jacobi Method\nIteration  x1old    x2old    x3old    x1new     x2new    x3new")
        for i in range(20):
            x1_j.append(-0.2*x2_j[i]-0.2*x3_j[i]+0.8)
            x2_j.append(-0.1*x1_j[i]-0.2*x3_j[i]-0.1)
            x3_j.append(-0.1*x1_j[i]-0.1*x2_j[i]-0.9)
            print("%02d        %7.4f  %7.4f  %7.4f  %7.4f   %7.4f  %7.4f" 
                  % (i+1,x1_j[i],x2_j[i],x3_j[i],x1_j[i+1],x2_j[i+1],x3_j[i+1]))
            if np.abs(x1_j[i+1]-1.) <= tol and np.abs(x2_j[i+1]) <= tol and np.abs(x3_j[i+1]+1.) <= tol:
                print('\n')
                break                
    if method_gs:
        x1_gs = [x1_i]
        x2_gs = [x2_i]
        x3_gs = [x3_i]
        print("Gauss-Seidel Method\nIteration  x1old    x2old    x3old    x1new     x2new    x3new")
        for i in range(20):
            x1_gs.append(-0.2*x2_gs[i]-0.2*x3_gs[i]+0.8)
            x2_gs.append(-0.1*x1_gs[i+1]-0.2*x3_gs[i]-0.1)
            x3_gs.append(-0.1*x1_gs[i+1]-0.1*x2_gs[i+1]-0.9)
            print("%02d        %7.4f  %7.4f  %7.4f  %7.4f   %7.4f  %7.4f" 
                  % (i+1,x1_gs[i],x2_gs[i],x3_gs[i],x1_gs[i+1],x2_gs[i+1],x3_gs[i+1]))
            if np.abs(x1_gs[i+1]-1.) <= tol and np.abs(x2_gs[i+1]) <= tol and np.abs(x3_gs[i+1]+1.) <= tol:
                print('\n')
                break
                
def iterative_3(x1_i=0, x2_i=0, x3_i=0, method_j=True, method_gs=False):
    tol = 5.e-5
    if method_j:
        x1_j = [x1_i]
        x2_j = [x2_i]
        x3_j = [x3_i]
        print("Jacobi Method\nIteration  x1old    x2old    x3old    x1new     x2new    x3new")
        for i in range(20):
            x1_j.append(0.4*x2_j[i]+0.6*x3_j[i]+0.8)
            x2_j.append(-0.2*x1_j[i]+0.8*x3_j[i]+1.5)
            x3_j.append(-0.2*x1_j[i]+0.2*x2_j[i]+1.23)
            print("%02d        %7.4f  %7.4f  %7.4f  %7.4f   %7.4f  %7.4f" 
                  % (i+1,x1_j[i],x2_j[i],x3_j[i],x1_j[i+1],x2_j[i+1],x3_j[i+1]))
            if np.abs(x1_j[i+1]-1.) <= tol and np.abs(x2_j[i+1]) <= tol and np.abs(x3_j[i+1]+1.) <= tol:
                print('\n')
                break                
    if method_gs:
        x1_gs = [x1_i]
        x2_gs = [x2_i]
        x3_gs = [x3_i]
        print("Gauss-Seidel Method\nIteration  x1old    x2old    x3old    x1new     x2new    x3new")
        for i in range(20):
            x1_gs.append(-0.2*x2_gs[i]-0.2*x3_gs[i]+0.8)
            x2_gs.append(-0.1*x1_gs[i+1]-0.2*x3_gs[i]-0.1)
            x3_gs.append(-0.1*x1_gs[i+1]-0.1*x2_gs[i+1]-0.9)
            print("%02d        %7.4f  %7.4f  %7.4f  %7.4f   %7.4f  %7.4f" 
                  % (i+1,x1_gs[i],x2_gs[i],x3_gs[i],x1_gs[i+1],x2_gs[i+1],x3_gs[i+1]))
            if np.abs(x1_gs[i+1]-1.) <= tol and np.abs(x2_gs[i+1]) <= tol and np.abs(x3_gs[i+1]+1.) <= tol:
                print('\n')
                break
```
