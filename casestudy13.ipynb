```r
install.packages("s20x")
library(s20x)
library(repr)
options(repr.plot.width=8, repr.plot.height=6)
```

```r
sleep.df = read.table("data/sleep.txt", header = TRUE)
head(sleep.df)
tail(sleep.df)
```

```r
boxplot(time ~ group, data = sleep.df)
```

```r
# fit the linear model
sleep.fit = lm(time ~ group, data = sleep.df)
```

```r
eovcheck(sleep.fit, levene = TRUE)
```

```r
normcheck(sleep.fit, shapiro.wilk = TRUE)
```

```r
summary(sleep.fit)
```
