# Lecture Demos - Fourier Series

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def fs_ex2a():
    x = [-6.*np.pi,-4.*np.pi,-2.*np.pi,0.,2*np.pi,4.*np.pi,6*np.pi]
    f = [6.*np.pi,0.,6.*np.pi,0.,6.*np.pi,0.,6.*np.pi]
    plotfs(x,f)

def fs_ex2b():
    x = np.linspace(-3.*np.pi+1e-4,3.*np.pi-1e-4,1001)
    f = np.sin(x)
    plotfs(x,f)

def fs_cos():
    x = np.linspace(-3.*np.pi+1e-4,3.*np.pi-1e-4,1001)
    f = np.cos(x)
    plotfs(x,f)
    
def fs_ex2c():
    x = [-3.,-2.,-2.,-1.,-1.,0.,0.,1.,1.,2.,2.,]
    f = [6.*np.pi,0.,6.*np.pi,0.,6.*np.pi,0.,6.*np.pi]
    
def plotfs(x,f):
    fig = plt.figure(figsize=(16, 8))
    ax1 = fig.add_subplot(111)
    ax1.set_xlabel('x')
    ax1.set_ylabel('f(x)')
    ax1.plot(x,f, 'k-')
    ax1.axhline(linewidth=0.5, color='k')
    ax1.axvline(linewidth=0.5, color='k')
    plt.show()
```

## Lecture Example 6A

```python
def fs_func(x,T):
    xc = x-np.rint(x/T)*T
    #print (x,xc,np.rint(x/T))
    if -1*np.pi <= xc <= -0.5*np.pi:
        return -1
    elif -0.5*np.pi <= xc <= 0.5*np.pi:
        return 1
    elif 0.5*np.pi <= xc <= 1*np.pi:
        return -1

n = 1001
x = np.linspace(-3*np.pi+0.001,3*np.pi-0.001,n)
f = np.zeros(n)
for i in range(n):
    f[i] = fs_func(x[i], 2.*np.pi)
sn = 5
s1 = 4.*np.cos(x)/(np.pi)
s3 = s1-4.*np.cos(3.*x)/(3.*np.pi)
s5 = s3+4.*np.cos(5.*x)/(5.*np.pi)
    
fig = plt.figure(figsize=(16, 8))
ax1 = fig.add_subplot(111)
ax1.set_xlabel('x')
ax1.set_ylabel('f(x)')
ax1.plot(x,f, 'k-')
ax1.plot([-9.,9.],[0.,0.], 'b-')
ax1.plot([0.,0.],[-1.1,1.1], 'b-')
#ax1.plot(x,s1, 'r--')
#ax1.plot(x,s3, 'g--')
ax1.plot(x,s5, 'm--')
```

## Lecture Example 6B

```python
def fs_func(x,T):
    xc = x-np.rint(x/T)*T
    if -1*np.pi <= xc <= -0.5*np.pi:
        return -1.*xc-np.pi
    elif -0.5*np.pi <= xc <= 0.5*np.pi:
        return xc
    elif 0.5*np.pi <= xc <= 1*np.pi:
        return -1.*xc+np.pi

n = 1001
x = np.linspace(-3*np.pi+0.001,3*np.pi-0.001,n)
f = np.zeros(n)
for i in range(n):
    f[i] = fs_func(x[i], 2.*np.pi)
sn = 5
s1 = 4.*np.sin(x)/(np.pi)
s3 = s1-4.*np.sin(3.*x)/(9.*np.pi)
s5 = s3+4.*np.sin(5.*x)/(25.*np.pi)
    
fig = plt.figure(figsize=(16, 8))
ax1 = fig.add_subplot(111)
ax1.set_xlabel('x')
ax1.set_ylabel('f(x)')
ax1.plot(x,f, 'k-')
ax1.plot([-10.,10.],[0.,0.], 'b-')
ax1.plot([0.,0.],[-1.7,1.7], 'b-')
#ax1.plot(x,s1, 'r--')
#ax1.plot(x,s3, 'g--')
ax1.plot(x,s5, 'm--')
```
