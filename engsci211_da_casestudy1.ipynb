# Caset Study 1. Crash Test Dummies: One Sample Analysis

This data comes from the National Transportation Safety Administration. It is based on trials in which stock automobiles were crashed into a wall at 35mph with dummies in the driver and front passenger seat. For each crash in each car data has been collected for the passenger and driver dummies. The data is ordered as follows: dummy in driver seat of car 1 to driver in dummy seat of car 112, then dummy in passenger seat of car 1 to dummy in passenger seat of car 112. The data is stored in crash.txt under the following variable headings:

\begin{align*}
\textbf{make} &\quad\quad \textrm{Car make} \\
\textbf{Model} &\quad\quad \textrm{Model of that car} \\
\textbf{HeadIC} &\quad\quad \textrm{Head injury criterion} \\
\textbf{Chestdecel} &\quad\quad \textrm{Chest deceleration} \\
\textbf{LLeg} &\quad\quad \textrm{Left femur load} \\
\textbf{RLeg} &\quad\quad \textrm{Right femur load} \\
\textbf{D.P} &\quad\quad \textrm{Dummy in driver or passenger seat.} \\
\textbf{Protection} &\quad\quad \textrm{Kind of protection (seatbelt, airbag etc.)} \\
\textbf{Doors} &\quad\quad \textrm{Number of doors on the car} \\
\textbf{Year} &\quad\quad \textrm{Year of the car} \\
\textbf{Wt} &\quad\quad \textrm{Weight in pounds} \\
\textbf{Size} &\quad\quad \textrm{Car classification type (light, minivan etc.)} \\
\end{align*}

Source: http://lib.stat.cmu.edu/DASL/Stories/CrashTestDummies.html

## Methods and Assumption Checks

We are attempting to estimate the mean of one group of observations, so
we do a one-sample *t*-test.

The weight of the cars should be independent of one another. The
Normality assumption appears to be satisfied as the points lie on the
straight line of the Q-Q plot.

The model fitted is ${\tt Wt}_i = \mu + \varepsilon_i$, where $\mu$
is the mean car weight and $\varepsilon_i \overset{\text{iid}}{\sim} N(0, \sigma^2)$.

## Executive Summary

Using data collected from the National Transportation Safety
Administration, we are interested in estimating the average weight of
the cars on the road. This data is based on 112 stock automobiles.

We estimate that the mean weight of the cars on the road is between 2686
and 2876 pounds.
